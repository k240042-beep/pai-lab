```python
def processTransactions(transactionList):
    customerData = {}
    for t in transactionList:
        cid = t['customerId']
        pid = t['productId']

        if cid not in customerData:
            customerData[cid] = set()

        customerData[cid].add(pid)

    return customerData
```

#2) Finding Frequent Pairs

```python
def findFrequentPairs(customerData):
    from itertools import combinations

    pairCount = {}

    for products in customerData.values():
        for a, b in combinations(sorted(products), 2):
            pair = frozenset([a, b])

            if pair not in pairCount:
                pairCount[pair] = 0

            pairCount[pair] += 1

    return pairCount
```

# 3)Get Recommendations

```python
def getRecommendations(targetProductId, frequentPairs):
    results = []

    for pair, cnt in frequentPairs.items():
        if targetProductId in pair:
            other = list(pair - {targetProductId})[0]
            results.append((other, cnt))

    results.sort(key=lambda x: x[1], reverse=True)
    return results
```

# 4)Generating Report

```python
def generateReport(targetProductId, recommendations, catalog):
    lines = []
    lines.append(f"Recommendations for product {targetProductId} ({catalog[targetProductId]}):\n")

    rank = 1
    for prod, cnt in recommendations:
        name = catalog[prod]
        lines.append(f"{rank}. {name} (productId: {prod}) â€“ co-purchased {cnt} times")
        rank += 1

    return "\n".join(lines)
```
