# Package Class


```python
class Package:
    def __init__(self, packageId, weightInKg):
        self.packageId = packageId
        self.weightInKg = weightInKg
```

# Drone Class

```python
class Drone:
    def __init__(self, droneId, maxLoadInKg):
        self.droneId = droneId
        self.maxLoadInKg = maxLoadInKg

        self.__status = "idle"
        self.currentPackage = None
```

**Getter**

```python
def getStatus(self):
        return self.__status
```

**setter**

```python
def setStatus(self, newStatus):
        if newStatus not in ["idle", "delivering", "charging"]:
            raise ValueError("Invalid drone status.")
        self.__status = newStatus
```

**Assign** **Package**

```python
 def assignPackage(self, packageObj):
        if self.__status != "idle":
            return False

        if packageObj.weightInKg > self.maxLoadInKg:
            return False

        self.currentPackage = packageObj
        return True
```

# Fleet Manager Class

```python
class FleetManager:
    def __init__(self):
        self.drones = {}
        self.pendingPackages = []

    def addDrone(self, droneObj):
        self.drones[droneObj.droneId] = droneObj

    def addPackage(self, packageObj):
        self.pendingPackages.append(packageObj)
```

**Assign jobs to idle drones**

```python
def dispatchJobs(self):
        for drone in self.drones.values():
            if drone.getStatus() == "idle" and len(self.pendingPackages) > 0:
                pkg = self.pendingPackages[0]

                if drone.assignPackage(pkg):
                    drone.setStatus("delivering")
                    self.pendingPackages.pop(0)
```

# Simulation Tick

```python
def simulationTick(self):
        for drone in self.drones.values():
            if not hasattr(drone, "stateTimer"):
                drone.stateTimer = 0

            drone.stateTimer += 1
            if drone.getStatus() == "delivering":
                if drone.stateTimer == 2:
                    drone.setStatus("charging")
                    drone.currentPackage = None
                    drone.stateTimer = 0

            elif drone.getStatus() == "charging":
                if drone.stateTimer == 2:
                    drone.setStatus("idle")
                    drone.stateTimer = 0
```
