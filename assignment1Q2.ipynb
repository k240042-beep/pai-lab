# 1)Cleaning Text

```python
def preprocessText(text, punctuationList, stopwordSet):
    t = text.lower()
    for p in punctuationList:
        t = t.replace(p, " ")
    words = t.split()

    cleaned = [w for w in words if w not in stopwordSet]

    return cleaned
```

# 2)Score Sentiment

```python
def analyzePosts(postsList, punctuation, stopwords, positive, negative):
    processed = list(
        map(
            lambda post: {
                "id": post["id"],
                "text": post["text"],
                "clean": preprocessText(post["text"], punctuation, stopwords),
                "score": sum(
                    [+1 if w in positive else -1 if w in negative else 0
                     for w in preprocessText(post["text"], punctuation, stopwords)]
                )},
            postsList
        ))
    return processed
```

# 3)Flag Negative Posts

```python
def getFlaggedPosts(scoredPosts, sentimentThreshold=-1):
    return [post for post in scoredPosts if post["score"] <= sentimentThreshold]
```

# 4)Finding Trending Negative Topics

```python
def findNegativeTopics(flaggedPosts):
    topicCount = {}

    for post in flaggedPosts:
        words = post["text"].split()

        for w in words:
            if w.startswith("#") or w.startswith("@"):
                topicCount[w] = topicCount.get(w, 0) + 1

    return topicCount
```
